# Explore Job Outcomes of College Graduates

### In this project, I will use the pandas plotting functionality to explore the data by visualizations.

_ Learning Note -- Pandas: pandas has many methods for quickly generating common plots from data in DataFrames. Like pyplot, the plotting functionality in pandas is a wrapper for matplotlib. This means we can customize the plots when necessary by accessing the underlying Figure, Axes, and other matplotlib objects._

We'll be working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012. The original data on job outcomes was released by American Community Survey, which conducts surveys and aggregates the data. The dataset can be found here: https://github.com/fivethirtyeight/data/tree/master/college-majors

Each row in the dataset represents a different major in college and contains information on gender diversity, employment rates, median salaries, and more. Here are some of the columns in the dataset:

Rank - Rank by median earnings (the dataset is ordered by this column).
Major_code - Major code.
Major - Major description.
Major_category - Category of major.
Total - Total number of people with major.
Sample_size - Sample size (unweighted) of full-time.
Men - Male graduates.
Women - Female graduates.
ShareWomen - Women as share of total.
Employed - Number employed.
Median - Median salary of full-time, year-round workers.
Low_wage_jobs - Number in low-wage service jobs.
Full_time - Number employed 35 hours or more.
Part_time - Number employed less than 35 hours.


### Set up the environment by importing the libraries, and use Jupyter magic to display plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

### Read the dataset into a DataFrame
iloc[] to return the first row formatted as a table;
head() and tail() to see how the data is structured
describe() to generate a summary statistics for all the numeric columns.

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
```

```python
print(recent_grads.head())
```

```python
print(recent_grads.tail())
```

```python
print(recent_grads.describe())
```

### Drop the missing values.

```python
recent_grads = recent_grads.dropna()
```

## Generate scatter plot to explore the data.
The relation between Sample size and Median salary. 

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
```

The relation between sample size and unemployment rate.

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
```

The relation between women share and the unemployment rate.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
```

The relation between number of genders and the median salary.

```python
recent_grads.plot(x='Men', y='Median', kind='scatter')
recent_grads.plot(x='Women', y='Median', kind='scatter')
```

### Do students in more popular majors make more money?
- Popular majors tend to make less money.

```python
recent_grads.plot(x='Total', y='Median', kind='scatter')
```

### Do students that majored in subjects that were majority female make more money?
- Subjects that were majority female tend to make less money.

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter')
```

### Is there any link between the number of fulltime employees and the median salary?
- Majors with more fulltime employees 

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter')
```

## Generate histograms and explore the distributions of the columns.
Default bins = 10. rot = angle of the labels.
- Most common median salary range is 30000-40000.
- More than half of the majors are predominantly female.

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", 
        "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(15,40))
for r in range(1,8):
    ax = fig.add_subplot(6,2,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
    ax.set_title(cols[r])
```

## Scatter matrix plot
Combine both scatter plots and histograms into one grid of plots and to explore potential relationships and distributions simultaneously.
Create a 2 by 2 scatter matrix with sample_size and Median salary.

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,6))
```

3 by 3 scatter matrix with sample_size, median and unemployment_rate.

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(15,15))
```

## Bar charts
Compare the percentages of women from the first 10 rows and last 10 rows of the sorted dataframe on unemployment rate.
- See the percentage of women in the most and least employment majors.

```python
sort_by_unemployment= recent_grads.sort_values('Unemployment_rate')
sort_by_unemployment[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
sort_by_unemployment[163:].plot.bar(x='Major', y='ShareWomen', legend=False)
```
